## Ages :

Tels que definis dans le formulaire et dans la fonction "get_age()" dans csv_OpportuniteGP.py

_As defined in the TypeForm questions and in "get_age()" function in csv_OpportuniteGP.py_

- Moins de 18 ans
- Entre 18 et 25 ans
- Entre 26 et 35 ans
- Entre 36 et 50 ans (valait en v1 / _used to be in v1_ : Entre 35 et 50 ans )
- Entre 51 et 75 ans
- 76 ans et plus

```python
data = []
tranches = []

# Conditions pour ajuster la legende au dessin (pas de legende si 0)
# Je sais qu'on n'est pas vieux a 25 ans. Je sais aussi qu'on n'est pas mega vieux a 75 ans. En gros on est ce qu'on est et moi j'ecris des commentaires pour rigoler
# Conditions to adjust label and data (not printing labels when 0)
mineurs, jeunes, vieux, pluvieux, supervieux, megavieux = dt.get_age()
if mineurs.shape[0] > 0 :
    data.append(mineurs.shape[0])
    tranches.append('mineurs')
if jeunes.shape[0] > 0 :
    data.append(jeunes.shape[0])
    tranches.append('[ 18 - 25 ]')
if vieux.shape[0] > 0 :
    data.append(vieux.shape[0])
    tranches.append('] 25 - 35 ]')
if pluvieux.shape[0] > 0 :
    data.append(pluvieux.shape[0])
    tranches.append('] 35 - 50 ]')
if supervieux.shape[0] > 0 :
    data.append(supervieux.shape[0])
    tranches.append('] 50 - 75 ]')
if megavieux.shape[0] > 0 :
    data.append(megavieux.shape[0])
    tranches.append('> 75')
    

# Creating plot
fig = plt.figure(figsize =(10, 7))
plt.pie(data, labels = tranches)
  
# show plot
plt.show()
```
